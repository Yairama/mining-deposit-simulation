```python
# importar librerías necesarias
import numpy as np

# función que implementa el algoritmo de Lerchs & Grossman en 3D
def lerchs_grossman_3D(block_data, mining_cost):
    # inicializar variables para almacenar las coordenadas xyz de cada bloque y si debe ser minado o no
    block_coordinates = []
    block_status = []
    final_economic_value = 0
    # obtener las coordenadas x, y, z y el valor económico de cada bloque
    for block in block_data:
        x, y, z, economic_value = block
        # determinar si el bloque debe ser minado o no utilizando la relación entre el valor económico y el costo de extracción
        if economic_value - mining_cost > 0:
            block_status.append(1)
            final_economic_value += economic_value - mining_cost
        else:
            block_status.append(0)
        block_coordinates.append([x, y, z])
    # retornar las coordenadas xyz y el estado de minería (minado o no) para cada bloque, y el valor económico final
    return block_coordinates, block_status, final_economic_value

# ejemplo de uso
block_data = [[1, 2, 3, 4], [2, 3, 4, 5], [3, 4, 5, 6], [4, 5, 6, 7]]
mining_cost = 1
block_coordinates, block_status, final_economic_value = lerchs_grossman_3D(block_data, mining_cost)
print("Coordenadas de los bloques:", block_coordinates)
print("Estatus de minería de los bloques:", block_status)
print("Valor económico final:", final_economic_value)
```
